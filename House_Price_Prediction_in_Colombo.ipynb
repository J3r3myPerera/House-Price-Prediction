```python
colombo_region = df1[df['Location'].str.contains('Colombo')]
colombo_region
```

```python
# Remove non-numeric characters and convert to float
# colombo_region['Price'] = colombo_region['Price'].str.replace('Rs ', '').str.replace(',', '').astype('float32')
colombo_region['Price'] = colombo_region['Price'].str.replace('Rs ', '').str.replace(',', '')
```

```python
column_name = 'Price'
data_type = df[column_name].dtype
data_type
```

```python
colombo_region.head()
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
X = colombo_region[['Baths','Beds']] #the features that I am using
y = colombo_region['Price'] # the target variable
```

```python
#for the normalization part
X_norm = (X - X.mean()) / X.std()
```

```python
X_norm = X_norm.astype('float32')
y = y.astype('float32')
```

```python
#building the model
model = keras.Sequential([
    keras.layers.Dense(32, activation = 'relu', input_shape =(2,)),
    keras.layers.Dense(32, activation= 'relu'),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_norm, y, epochs=100, batch_size=32)
```

```python
test_data = pd.DataFrame({'bedrooms': [3, 4], 'bathrooms': [2, 3]})
normalized_test_data = (test_data - X.mean()) / X.std()
predictions = model.predict(normalized_test_data)
```

```python
predictions = predictions.astype('float64')
print(predictions)
```

```python
# Calculate the errors (difference between predictions and actual prices)
errors = y - predictions.flatten()
```
